```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
hospital_readmissions_train = pd.read_csv('hospital_readmissions_train.csv')

# Display the first few rows of the dataset
print(hospital_readmissions_train.head())

# Get an overview of the dataset
print(hospital_readmissions_train.info())

# Check for missing values
print(hospital_readmissions_train.isnull().sum())

# Check for duplicate rows
print(hospital_readmissions_train.duplicated().sum())

# Exploratory Data Analysis

# Target variable distribution
plt.figure(figsize=(10, 6))
sns.countplot(x='Stay', data=hospital_readmissions_train)
plt.title('Distribution of Length of Stay')
plt.xlabel('Length of Stay')
plt.ylabel('Count')
plt.show()

# Distribution of numerical variables
num_cols = ['Available Extra Rooms in Hospital', 'Visitors with Patient', 'Admission_Deposit']
for col in num_cols:
    plt.figure(figsize=(10, 6))
    sns.histplot(hospital_readmissions_train[col], bins=20, kde=True)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.show()

# Correlation analysis (excluding categorical columns)
num_cols = ['Available Extra Rooms in Hospital', 'Visitors with Patient', 'Admission_Deposit']
correlation_matrix = hospital_readmissions_train[num_cols].corr()

plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix for Numerical Variables')
plt.show()

# Heatmap for all numerical columns
plt.figure(figsize=(12, 8))
sns.heatmap(hospital_readmissions_train.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix for All Numerical Variables')
plt.show()

# Conclusion
# Summarize key findings and insights from the EDA
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
hospital_readmissions_train = pd.read_csv('hospital_readmissions_train.csv')

# Select only numeric columns for correlation analysis
numeric_cols = hospital_readmissions_train.select_dtypes(include=['number'])

# Correlation analysis
plt.figure(figsize=(12, 8))
sns.heatmap(numeric_cols.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix for All Numerical Variables')
plt.show()
```
