```python
import numpy as mp
import pandas as pd
import seaborn as sns
import plotly.express as pe
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("oilgas.csv")
```

```python
data
```

```python
numeric_columns = ['azimuth', 'md (ft)', 'tvd (ft)', 'footage lateral length', 'well spacing', 'porpoise deviation', 'porpoise count', 'shale footage', 'acoustic impedance', 'log permeability', 'porosity', 'poisson ratio', 'water saturation', 'toc', 'vcl', 'p-velocity', 's-velocity', 'youngs modulus', 'isip', 'breakdown pressure', 'pump rate', 'total number of stages', 'proppant volume', 'proppant fluid ratio', 'production']
data[numeric_columns] = data[numeric_columns].fillna(data[numeric_columns].mean())
```

```python
data.isnull().sum()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
plt.hist(data['shale footage'], bins=20, color='blue', alpha=0.7, label='Shale Footage')
plt.xlabel('Shale Footage')
plt.ylabel('Frequency')
plt.title('Histogram of Shale Footage')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
plt.scatter(data['shale footage'], data['production'], alpha=0.7, color='blue')
plt.xlabel('Shale Footage')
plt.ylabel('Production')
plt.title('Scatter Plot: Shale Footage vs. Production')
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.scatter(data['shale footage'], data['production'], alpha=0.7)
plt.xlabel('Shale Footage')
plt.ylabel('Production')
plt.title('Scatter Plot: Shale Footage vs. Production')
plt.show()
```

```python
correlation = data['shale footage'].corr(data['production'])
print(f'Correlation Coefficient: {correlation}')
```

```python
import seaborn as sns
import statsmodels.api as sm
sns.regplot(x='shale footage', y='production', data=data, scatter_kws={'alpha':0.7})
plt.title('Regression Analysis: Shale Footage vs. Production')
X = sm.add_constant(data['shale footage'])
model = sm.OLS(data['production'], X).fit()
print(model.summary())
```
