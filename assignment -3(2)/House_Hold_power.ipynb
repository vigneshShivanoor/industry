```python
da = pd.read_csv('household_power_consumption.txt', sep=';', parse_dates={'datetime': ['Date', 'Time']}, infer_datetime_format=True)
```

```python
da
```

```python
da.info()
```

```python
da['Global_active_power'].fillna(int(da['Global_active_power'].mean()),inplace=True)
```

```python
da['Sub_metering_1'] = pd.to_numeric(da['Sub_metering_1'], errors='coerce')
daily_sub_metering_1 = da.groupby(da['datetime'].dt.date)['Sub_metering_1'].sum()
plt.figure(figsize=(12, 6))
plt.plot(daily_sub_metering_1.index, daily_sub_metering_1.values, marker='o', linestyle='-', color='b')
plt.title('Daily Sub Metering 1')
plt.xlabel('Date')
plt.ylabel('Sub Metering 1 Value')
plt.grid(True)
plt.show()
```

```python
da['Sub_metering_2'] = pd.to_numeric(da['Sub_metering_2'], errors='coerce')
daily_sub_metering_1 = da.groupby(da['datetime'].dt.date)['Sub_metering_2'].sum()
plt.figure(figsize=(12, 6))
plt.plot(daily_sub_metering_1.index, daily_sub_metering_1.values, marker='o', linestyle='-', color='b')
plt.title('Daily Sub Metering 2')
plt.xlabel('Date')
plt.ylabel('Sub Metering 2 Value')
plt.grid(True)
plt.show()
```

```python
da['Sub_metering_3'] = pd.to_numeric(da['Sub_metering_3'], errors='coerce')
daily_sub_metering_1 = da.groupby(da['datetime'].dt.date)['Sub_metering_3'].sum()
plt.figure(figsize=(12, 6))
plt.plot(daily_sub_metering_1.index, daily_sub_metering_1.values, marker='o', linestyle='-', color='b')
plt.title('Daily Sub Metering 3')
plt.xlabel('Date')
plt.ylabel('Sub Metering 3 Value')
plt.grid(True)
plt.show()
```

```python
da['Year'] = da['datetime'].dt.year
yearly_sub_metering_1 = da.groupby('Year')['Sub_metering_1'].sum()
plt.figure(figsize=(12, 6))
plt.plot(yearly_sub_metering_1.index, yearly_sub_metering_1.values, marker='o', linestyle='-', color='b')
plt.title('Yearly Sub Metering 1')
plt.xlabel('Year')
plt.ylabel('Sub Metering 1 Value (Yearly Sum)')
plt.grid(True)
plt.show()
```
